```python
# =========================================
# SIMULADOR DE IA PARA ILUSTRADOR DIGITAL
# Uso de listas, if, elif, else y while True
# Nivel: Preparatoria
# =========================================

# Lista vacía al inicio
ilustrador = []

print("🎨 ====================================")
print("   IA PARA ILUSTRADOR DIGITAL")
print("==================================== 🎨")

while True:

    print("\n--------- MENÚ PRINCIPAL ---------")
    print("1. Agregar dibujo")
    print("2. Ver dibujos guardados")
    print("3. Buscar dibujo")
    print("4. Eliminar dibujo")
    print("5. Recomendación de IA")
    print("6. Salir")

    opcion = input("Selecciona una opción: ")

    # ====================================
    # AGREGAR DIBUJO
    # ====================================
    if opcion == "1":

        nombre = input("Nombre del dibujo: ")
        estilo = input("Estilo (anime, realista, pixel art, etc): ")
        color = input("Color principal: ")

        dibujo = [nombre, estilo, color]

        ilustrador.append(dibujo)

        print("\n✅ Dibujo agregado correctamente")
        print("La IA guardó tu ilustración 🎨")

    # ====================================
    # MOSTRAR DIBUJOS
    # ====================================
    elif opcion == "2":

        if ilustrador == []:
            print("\n⚠ La lista está vacía")
            print("No hay dibujos registrados")

        else:
            print("\n📚 DIBUJOS GUARDADOS")

            posicion = 0

            while posicion < len(ilustrador):

                print("\nDibujo", posicion + 1)
                print("Nombre:", ilustrador[posicion][0])
                print("Estilo:", ilustrador[posicion][1])
                print("Color:", ilustrador[posicion][2])

                posicion = posicion + 1

    # ====================================
    # BUSCAR DIBUJO
    # ====================================
    elif opcion == "3":

        buscar = input("Escribe el nombre del dibujo: ")

        posicion = 0
        encontrado = False

        while posicion < len(ilustrador):

            if buscar == ilustrador[posicion][0]:

                print("\n🔍 Dibujo encontrado")
                print("Nombre:", ilustrador[posicion][0])
                print("Estilo:", ilustrador[posicion][1])
                print("Color:", ilustrador[posicion][2])

                encontrado = True

            posicion = posicion + 1

        if encontrado == False:
            print("\n❌ La IA no encontró ese dibujo")

    # ====================================
    # ELIMINAR DIBUJO
    # ====================================
    elif opcion == "4":

        eliminar = input("Nombre del dibujo a eliminar: ")

        posicion = 0
        eliminado = False

        while posicion < len(ilustrador):

            if eliminar == ilustrador[posicion][0]:

                ilustrador.pop(posicion)

                print("\n🗑 Dibujo eliminado correctamente")

                eliminado = True

            else:
                posicion = posicion + 1

        if eliminado == False:
            print("\n❌ No existe ese dibujo")

    # ====================================
    # RECOMENDACIÓN DE IA
    # ====================================
    elif opcion == "5":

        estilo = input("¿Qué estilo te gusta?: ")

        if estilo == "anime":
            print("\n🤖 IA recomienda:")
            print("Usar colores brillantes y ojos grandes")

        elif estilo == "realista":
            print("\n🤖 IA recomienda:")
            print("Agregar sombras y detalles en la piel")

        elif estilo == "pixel art":
            print("\n🤖 IA recomienda:")
            print("Utilizar pocos colores y cuadros pequeños")

        else:
            print("\n🤖 IA recomienda practicar más técnicas de dibujo")

    # ====================================
    # SALIR
    # ====================================
    elif opcion == "6":

        print("\n👋 Cerrando el sistema de IA...")
        print("Gracias por usar el ilustrador digital")
        break

    # ====================================
    # OPCIÓN INVÁLIDA
    # ====================================
    else:

        print("\n⚠ Opción no válida")
```
